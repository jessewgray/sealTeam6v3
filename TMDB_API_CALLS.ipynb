```python
import pandas as pd

import requests 
from pprint import pprint

oscar_movie_db = pd.read_csv('./DataSources/Oscars1934-2015.csv')
oscar_movie_db.head()
```

```python
best_pictures_df = oscar_movie_db[((oscar_movie_db['Award']== 'Best Picture')|(oscar_movie_db['Award']== 'Best Motion Picture'))]

best_pictures_df.head()
```

```python
best_pictures_df['Name']

best_picture_noms = []
for name in best_pictures_df['Name']:
    best_picture_noms.append(name)
best_picture_noms
```

```python
# use api.tmdb.org due to problem of 'query limit is 40'
base_url ='https://api.tmdb.org/3/search/movie?'
api_key = 'api_key=d5b4e5d5145be601d4640d76efd791e3'
query = '&query='
query_base_url = base_url + api_key + query

# best_pic_info: get the movie summary information 
best_pic_info = []
picNumList = []
for i in range(len(best_picture_noms)): 
    try:
        release_year = best_picture_noms[i],'Year'.item()
        #print(release_year)
    except:
        release_year = 9999
    movie_info = query_base_url + best_picture_noms[i] + '&year=' + str(release_year)
    response=requests.get(movie_info).json()
    best_pic_info.append(response['results'])
    try:
        picNumList.append(best_pic_info[i][0]['id'])
    except:
        picNumList.append(996)
picNumList
```

```python
response['results'][0]
```

```python
base_url1 = 'https://api.tmdb.org/3/movie/'

language = '&language=en-US'
Winner_Details= []

for i in range(len(picNumList)):
    tmdb_id = str(picNumList[i]) + '?'
    detail_query_link = base_url1 + tmdb_id + api_key + language
    response = requests.get(detail_query_link).json()
    Winner_Details.append(response)
print("done")
```

```python
title_list = []
company_list = []
revenue_list = []
budget_list = []
runtime_list = []
vote_average_list = []
vote_count_list =[]
pop_list = []
genre_list =[]
release_date_list = []
movie_id_list = []

for i in range(len(Winner_Details)):
    title_list.append(Winner_Details[i]['original_title'])
    try:
        company_list.append(Winner_Details[i]['production_companies'][0]['name'])
    except:
        company_list.append('N/A')
    revenue_list.append(Winner_Details[i]['revenue'])
    budget_list.append(Winner_Details[i]['budget'])
    runtime_list.append(Winner_Details[i]['runtime'])
    vote_average_list.append(Winner_Details[i]['vote_average'])
    vote_count_list.append(Winner_Details[i]['vote_count'])
    pop_list.append(Winner_Details[i]['popularity'])
    genre_list.append(Winner_Details[i]['genres'])
    release_date_list.append(Winner_Details[i]['release_date'])
    movie_id_list.append(Winner_Details[i]['id'])
    
print("done")

```

```python
newDB = {
    #'Year': best_pictures_df['Year'],
    #'Ceremony': best_pictures_df['Ceremony'],
    #'Award': best_pictures_df['Award'],
    #'Winner': best_pictures_df['Winner'],
    'Title': title_list,
    'Company': company_list,
    'Revenue': revenue_list,
    'Budget': budget_list,
    'Runtime': runtime_list,
    'Vote Average': vote_average_list,
    'Vote Count': vote_count_list, 
    'Popularity': pop_list,
    'Genre': genre_list,
    'Release Date': release_date_list,
    'Movie ID': movie_id_list
}
```

```python
newDF = pd.DataFrame(newDB)
newNewDF = newDF.drop_duplicates(subset=['Title'], keep='first')
newNewDF.shape
```
